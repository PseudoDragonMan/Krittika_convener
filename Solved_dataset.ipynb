# Python Assignment - Working With Events

In this question, you are given 4 text files. 
These files contain timestamps, which correspond to the times of some 'events'. 

Your first task is to find out how many events are present in each file.

Consider only those events which have a positive value for timestamp


```python
import matplotlib.pyplot as plt      #importing the necessary libraries.
import numpy as np
 
    #defining a function to extract data.

def get_data(filename):              
    with open(filename, 'r') as file:
        file_data = file.read().split(',')
        file_data = [float(x) for x in file_data if float(x) > 0]
        return file_data              
  #the function returns the data in a list.

filenames = ['event_Q1.txt', 'event_Q2.txt', 'event_Q3.txt', 'event_Q4.txt']    

   #storing the data from all the files in a multi-dimensional array.
file_data = []
for filename in filenames:         
    temp = get_data(filename)
    file_data.append(temp)
    print(f"The number of events with positive timestamp in {filename} is {len(temp)}")
   #note that ken is the number of files, basically
ken = len(file_data)
```

After this, for each file, make a plot of events per second in that file.
You dont have to find the average rate, instead for each second 0-1, 1-2, 2-3, .. 54-55, find the number of events in each second and plot it with respect to time.

Plot these 4 separately, as well as in a single overlapping plot.
Make sure to give proper titles, labels and legends.

```python
#defining a function to find the maximum of each set
def find_max(data):
    maxie = 0
    for i in range(len(data)):
        if data[i]>maxie:
            maxie = data[i]
    return maxie
#defining a function to get the frequency of events in each time timestamp.
#here, the timestamp 0-1 is represented by 1; 1-2 by 2 and so on.
def get_frequency(file_data, max_val):
    kein = int(np.ceil(max_val))
    freq = [0 for i in range(kein)]
    for i in range(len(file_data)):
        freq[int(np.floor(file_data[i]))] += 1
    return freq
#here, we make a multidimensional array of frequency, an array of maximum value and a multi-D array of timestamps.
max_val = []
frequency = []
time_stamp=[]
#using plot functions from matplotlib to plot a frequency vs timestamp graph
for i in range(len(file_data)):
    max_val.append(find_max(file_data[i]))
    frequency.append(get_frequency(file_data[i], max_val[i]))
    x = np.arange(0, int(np.ceil(max_val[i])), 1)
    time_stamp.append(x+1)
    plt.plot(x, frequency[i], label=filenames[i])
    plt.title(filenames[i])
    plt.xlabel('Time stamps')
    plt.ylabel('Frequency of events')
    plt.legend()
    plt.show()
```

Now for the final part of this question. We can see that the event frequency varies with time. Count how many bins are there, where atleast 2 of the files have more than 140 events in that bin, simultaneously.

```python
#defining a function to check if the given time stamp has the properties of a favorable bin wrt parameters num_files and num_bins
common_bins = 0
def check_bins(time_stp, num_bins, num_files):
    count =0
    for i in range(len(file_data)):
        if time_stp < len(time_stamp[i]):
            if frequency[i][time_stp-1] > num_bins:
                count += 1
        if count >= num_files:
            return True
    return False
#finding max timestamp
max_ts = max(max(time_stamp[i]) for i in range(4))
for i in range(1, max_ts+1):
    if check_bins(i, 140, 2):
        common_bins += 1
print(common_bins)
```
