```python
#importing the necessary modules and functions.
import requests      
from bs4 import BeautifulSoup  # We'll be working on webscraping using this library.
import pandas as pd            # Pandas is the go-to library for data science. We'll just be using it for making a csv.

page = requests.get("https://krittikaiitb.github.io/team/")
```

```python
page.content   #This function prints the page.
```

```python
soup = BeautifulSoup(page.content, 'html.parser')  
```

```python
readable = soup.prettify() #This function here, makes the code more readable. Run the output code as print('code') to get a more beatified verision. Makes it very easy to read.
readable
```

```python
dahtah = [] #as you can see, I pronounce data as data. Clearly, a superior kind. Wot else y'all need in a convenr? :D
year = []
person = []
post = []

for i in range(4):
     
        #web scraping by isolating text using classes and id

        #I've used the only repetitive condition available across all the years
    daytuh = soup.find_all('div', attrs = {'class': "card shadow-lg container mt-4 mb-5 aboutCard"})[i]
        #Isolate the name
    for element in daytuh.find_all('h5', attrs = {'class':'card-title mb-0'}):
            person.append(element.get_text())
            year.append(2022-i)
            #Isolate the class
    for element in daytuh.find_all('p', attrs = {'class':"card-text"}):
        post.append(element.get_text())

dahtah.append(person)
dahtah.append(post)
dahtah.append(year)


df = pd.DataFrame(person, columns= ['Names'])  #This is whwere we convert it into a data frame.
df['Designation']=post
df['Year of office']=year

#Enter your file address in the following format to save it the required location -> df.to_csv(r'file location', index= false)

df.to_csv('Krittika.csv', index=False)
read = pd.read_csv('Krittika.csv')
print(read)
#                   /\__/\
#AND WE'RE DONE    (='.'=)
#                    _._
```
